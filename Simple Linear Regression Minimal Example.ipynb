```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate Random Data to Train With 
observations = 1000
# np.random.uniform(low,high,size) -> draws a random value from the interval (low,high) where each number has an equal
# change of being selected
xs = np.random.uniform(low=-10,high=10,size = (observations,1))
zs = np.random.uniform(low=-10,high=10,size = (observations,1))
inputs = np.column_stack((xs,zs))

print(inputs.shape)
```

```python
# Create the targets we will aim at
noise = np.random.uniform(-1,1,size = (observations,1))
targets = 2*xs - 3*zs + 5 + noise 
print(targets.shape)
```

```python
# Plot the Training Data
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim = 200)
plt.show()
targets = targets.reshape(observations,1)
```

```python
# Initialize Variables
init_range = 0.1
# that will be the radius of the range we will use to initialize the weights and biases
weights = np.random.uniform(-init_range,init_range,size=(2,1)) # Because weights a 2(row) by 1 (column) matrix
# 2 weights for one output
biases = np.random.uniform(-init_range,init_range,size=1) # when just one cell/known as a scalar
print(weights)
print(biases)
```

```python
#Set a learning Rate
learning_rate = 0.02
```

```python
# Train the model
 # Iteration -> calculate outputs, compare outputs to targets through loss, print loss, adjust weights and biases
for i in range (100):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    loss = np.sum(deltas**2) / 2 / observations
    
    print(loss)
    
    deltas_scaled = deltas/observations
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights,biases)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
